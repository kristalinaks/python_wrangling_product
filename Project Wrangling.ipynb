# Melihat Data Yang Tersedia

### Melihat daftar tabel yang ada

```python
# loading in modules
import sqlite3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# creating file path
dbfile = 'olist.db'
# Create a SQL connection to our SQLite database
con = sqlite3.connect(dbfile)

# creating cursor
cur = con.cursor()

# reading all table names
table_list = [a for a in cur.execute("SELECT name FROM sqlite_master WHERE type = 'table'")]
# here is you table list
print(table_list)
```

```python
df_translation[df_translation.duplicated(keep=False)]
```

### Tabel Translation: Checking inconsistency (kolom product_category_name_english)

```python
df_translation["product_category_name_english"].unique()
```

```python
# Mengganti beberapa inkonsistensi
perubahan = {"home_confort": "home_comfort", "home_comfort_2": "home_comfort", "home_appliances_2":"home_appliances"}
df_translation = df_translation.replace(perubahan)

df_translation["product_category_name_english"].unique()
```

## Menggabungkan data

```python
# Menggabungkan tabel items, products, translation 
df1 = pd.merge(df_items, df_products, on="product_id", how="left")
df1 = pd.merge(df1, df_translation, on="product_category_name", how="left")

df1.head()
```

```python
# Cek null dari tabel gabungan
df1.isna().sum()
```

```python
# Cek kategori produk yang ada nama Portugisnya tapi tidak ada nama Inggrisnya
df1[["product_category_name","product_category_name_english"]].loc[
    (df1["product_category_name"].notna()) & (df1["product_category_name_english"].isna())]
```

```python
# Mengisi null di product_category_name_english
isian = {"pc_gamer":"pc_gamer", 
         "portateis_cozinha_e_preparadores_de_alimentos":"kitchen_and_food_preparators_portables", None:"others"}
df1["product_category_name_english"] = df1["product_category_name_english"].fillna(df1["product_category_name"].map(isian))

df1.isna().sum()
```

```python
# Menghapus kolom, mengganti nama kolom
df1_summary = df1.rename({"product_category_name_english":"product_category", 
                  "price":"sales"}, axis=1)

# Hitung sales per kategori
df1_summary = df1_summary[["product_category","sales"]].groupby(
    "product_category").sum().sort_values(
    "sales", ascending=False).reset_index()
df1_summary
```

### Visualisasi 1

```python
f, ax = plt.subplots(figsize=(10, 15))

sns.barplot(y="product_category", x="sales", data=df1_summary, color="b")
sns.set_style("whitegrid")

ax.set(xlim=(0, 1_300_000), ylabel="Product Category", xlabel="Total Sales", title="Total Penjualan dari Tiap Kategori Produk")

for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)
```

# Objektif 2: Kategori Produk yang Paling Banyak Dipesan

### Meringkas data

```python
# Menggunakan kembali tabel gabungan df1
df2 = df1.rename({"product_category_name_english":"product_category", "order_id":"order_counts"}, axis=1)
df2 = df2[["product_category","order_counts"]].groupby("product_category").count()
df2 = df2.sort_values("order_counts", ascending=False).reset_index()

df2
```

### Visualisasi 2

```python
f, ax = plt.subplots(figsize=(10, 15))

sns.barplot(y="product_category", x="order_counts", data=df2, color="g")
sns.set_style("whitegrid")

ax.set(xlim=(0, 14_000), ylabel="Product Category", xlabel="Total Order", title="Kategori Produk yang Paling Banyak Dipesan")

for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)
```

# Objektif 3: Pertumbuhan Pemesanan 10 Produk yang Paling Diminati di Tahun 2018

## Menyiapkan tabel

### Tabel Order: Checking null (memastikan kolom yang dibutuhkan tidak null)

```python
df_order.isna().sum()
```

### Tabel Order: Checking duplicated

```python
df_order[df_order.duplicated(keep=False)]
```

### Tabel Order: Meringkas tabel

```python
# Menghapus kolom yg tidak dibutuhkan
df_order = df_order.drop(columns=["customer_id","order_status","order_approved_at","order_delivered_carrier_date","order_delivered_customer_date","order_estimated_delivery_date"])

# Mengolah data tanggal
df_order["order_purchase_timestamp"] = pd.to_datetime(df_order["order_purchase_timestamp"], yearfirst=True)
df_order["month_purchased"] = df_order["order_purchase_timestamp"].dt.month
df_order["year_purchased"] = df_order["order_purchase_timestamp"].dt.year

df_order = df_order.loc[df_order["year_purchased"] == 2018]
df_order
```

## Menggabungkan data

```python
# Menggabungkan tabel order dengan tabel gabungan di objektif 1
df3 = pd.merge(df1, df_order, on="order_id", how="inner")

# Menggabungkan dengan data kategori produk dengan pemesanan terbanyak dari objektif 2
df2_top10 = df2.head(10)
df3 = df3.rename({"product_category_name_english":"product_category"}, axis=1).reset_index()
df3 = pd.merge(df2_top10[["product_category"]], df3[["product_category", "order_item_id", "month_purchased"]], on="product_category", how="left")
df3
```

```python
# Meringkas tabel menjadi pivot table
df3 = pd.pivot_table(df3.reset_index(), columns="product_category", index="month_purchased", values="order_item_id", aggfunc="sum")
df3
```

## Visualisasi 3

```python
ax = df3.plot(kind="line", title="Pertumbuhan Pemesanan 10 Produk yang Paling Diminati", xlabel="Bulan", ylabel="Banyak Order", figsize=(15,10))

ax.legend(title="Product Category")

for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)
```
